```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def Sin (t):
    return np.sin (2 * np.pi * t)
```

```python
def f (t):
    return 2 * Sin (2 * t) + Sin (20 * t)
```

```python
n = 1000
t = np.linspace (0, 2, n)
x = f (t)
pl.plot (x)
```

```python
z1 = np.fft.rfft (x)
x1 = np.real (z1)

pl.figure (figsize = (15, 5))
pl.subplot (1, 2, 1); pl.plot (x)
pl.subplot (1, 2, 2); pl.plot (x1)

#Frequenzen extrahieren: links das Signal und rechts die verschiedenen mglichen Frequenzen
```

```python
z1i = np.fft.irfft (z1)
x1i = np.real (z1i)

pl.figure (figsize = (15, 5))
pl.subplot (1, 2, 1); pl.plot (x1i)
pl.subplot (1, 2, 2); pl.plot (x - x1i)
```

```python
z2 = np.copy (z1)
z2 [10:] = 0+0j
x2 = np.real (z2)

z2i = np.fft.irfft (z2)
x2i = np.real (z2i)

pl.figure (figsize = (15, 5))
pl.subplot (1, 2, 1); pl.plot (x1, 'gray'); pl.plot (x2, 'red')
pl.subplot (1, 2, 2); pl.plot (x, 'gray'); pl.plot (x2i, 'red')


#Rotes signal ist das geglettete Signal welches wir erhalten wenn wir die hohen Frequenzen lschen
```

```python
z3 = np.copy (z1)
z3 [:10] = 0+0j
x3 = np.real (z3)

z3i = np.fft.irfft (z3)
x3i = np.real (z3i)

pl.figure (figsize = (15, 5))
pl.subplot (1, 2, 1); pl.plot (x1, 'gray'); pl.plot (x3, 'red')
pl.subplot (1, 2, 2); pl.plot (x, 'gray'); pl.plot (x3i, 'red')

#Tiefe Frequenz gelscht welches es dann auch der hohen Frequenz ermglicht gerade zu bleiben anstatt sich an der tiefen Frequenz zu orientieren
```

```python
def Gaussian (sigma, x):
    return np.exp (- x ** 2 / (2 * sigma ** 2)) / (sigma * np.sqrt (2 * np.pi))
```

```python
sigma = 5
r = 10
kernel = Gaussian (sigma, np.linspace (-r, r, 2 * r + 1))
kernel /= np.sum (kernel)
pl.stem (kernel)
```

```python
y = np.convolve (x, kernel, mode = 'same')
```

```python
pl.plot (x, 'gray')
pl.plot (y, 'red')

#Filter fr die glttung des Signals (Die Peaks sind klar kleiner aber der Verlust ist geringer als beim kompletten lschen der hohen Frequenzen)
```

```python
k = np.zeros_like (x)
k [n // 2 - r : n // 2 + r + 1] = kernel
pl.plot (x, 'gray')
pl.plot (k, 'blue')
```

```python
z4 = np.fft.ifft (np.fft.rfft (x) * np.fft.rfft (k))
x4 = np.real (z4)

print ('x:', x.shape)
print ('k:', k.shape)
print ('z4:', z4.shape)
print ('x4:', x4.shape)

pl.figure (figsize = (15, 5))
pl.subplot (1, 2, 1)
pl.plot (x [::2], 'gray')
pl.plot (y [::2], 'blue')
pl.plot (x4, 'red')
pl.subplot (1, 2, 2)
pl.plot (y [::2], 'blue')
pl.plot (x4, 'red')
pl.plot (y [::2] - x4 [:-1], 'red')
```
